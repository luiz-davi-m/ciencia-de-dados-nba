```python
from nba_api_jogadores_utils import buscar_id_jogador, obter_info_basica_jogador
from constantes_jogadores import JOGADORES, RESULTADOS_PATH
import pandas as pd
from nba_api_utils import salvar_dataset_csv
```

# RF1 -  Apresentar os seguintes dados dos Jogadores [Dependendo de “onde esteja extraindo” os dados, existe um Id, exiba também o Id de cada Jogador] (Vide Tabela 1). Se o jogador for um Rookie, converter para o número 0.

```python
def formatar_dados_jogador(nome_jogador):
    id_jogador = buscar_id_jogador(nome_jogador)
    
    if not id_jogador:
        return None
    
    info_df = obter_info_basica_jogador(id_jogador)
    
    dados_jogador = {
        'Nome': nome_jogador,
        'Altura': info_df['HEIGHT'].values[0] if 'HEIGHT' in info_df else None,
        'Peso': info_df['WEIGHT'].values[0] if 'WEIGHT' in info_df else None,
        'Idade': info_df['AGE'].values[0] if 'AGE' in info_df else None,
        'Experiência': info_df['SEASON_EXP'].values[0] if 'SEASON_EXP' in info_df else None,
        'Posição': info_df['POSITION'].values[0] if 'POSITION' in info_df else None,
        'Universidade': info_df['COLLEGE'].values[0] if 'COLLEGE' in info_df else None,
        'Salário': info_df['SALARY'].values[0] if 'SALARY' in info_df else None
    }
    
    return dados_jogador
```

```python
def retornar_dados_jogadores(jogadores):
    lista_dados_jogadores = [formatar_dados_jogador(jogador) for jogador in jogadores]
    lista_dados_jogadores = [dados for dados in lista_dados_jogadores if dados]
    
    return pd.DataFrame(lista_dados_jogadores)
```

```python
dados_jogadores = retornar_dados_jogadores(JOGADORES)
dados_jogadores
```

```python
dados_jogadores.loc[dados_jogadores["Nome"] == "Nikola Jokić", "Salário"] = "51,42 milhões USD"
dados_jogadores.loc[dados_jogadores["Nome"] == "Michael Porter Jr.", "Salário"] = "35,08 milhões USD"
dados_jogadores.loc[dados_jogadores["Nome"] == "Jamal Murray", "Salário"] = "36,02 milhões USD"

dados_jogadores
```
